```python
#aggregated_insurance in states
path0="D:/GUVI/Capstone Project/Phonepe_project/pulse/data/aggregated/insurance/country/india/state/"
ag_ins_li=os.listdir(path0)

#append data on column for dataframe
col0={"states":[],"year":[],"quarter":[],"transaction_type":[],"transaction_count":[],"transaction_amount":[]}

#for loop for states
for state in ag_ins_li:
    current_states=path0+state+"/"
    ag_year_li=os.listdir(current_states)
#for loop for years   
    for year in ag_year_li:
        current_year=current_states+year+"/"
        ag_json_list=os.listdir(current_year)
#for loop for all files        
        for file in ag_json_list:
            current_file=current_year+file
            data=open(current_file,"r")

#load data in json 
            phonedata0=json.load(data)

#Phonedata dataframe for aggregated transactions
            for i in phonedata0["data"]["transactionData"]:
                name=i["name"]
                count=i["paymentInstruments"][0]["count"]
                amount=i["paymentInstruments"][0]["amount"]
                col0["transaction_type"].append(name)
                col0["transaction_count"].append(count)
                col0["transaction_amount"].append(amount)
                col0["states"].append(state)
                col0["year"].append(year)
                col0["quarter"].append(int(file.strip(".json")))

#aggregated_insurance data into dataframe
aggregated_insurance=pd.DataFrame(col0)

#preprocessing
aggregated_insurance["states"]=aggregated_insurance["states"].str.replace('andaman-&-nicobar-islands','Andaman & Nicobar')
aggregated_insurance["states"]=aggregated_insurance["states"].str.replace("-","")
aggregated_insurance["states"]=aggregated_insurance["states"].str.title()
aggregated_insurance["states"]=aggregated_insurance["states"].str.replace("Dadra&Nagarhaveli&Daman&Diu","Dadra and Nagar Haveli and Daman and Di")
aggregated_insurance["states"]=aggregated_insurance["states"].str.replace("Jammu&Kashmir","Jammu & Kashmir")
aggregated_insurance["states"]=aggregated_insurance["states"].str.replace("Andhrapradesh","Andhra Pradesh")
aggregated_insurance["states"]=aggregated_insurance["states"].str.replace("Arunachalpradesh","Arunachal Pradesh")
aggregated_insurance["states"]=aggregated_insurance["states"].str.replace("Himachalpradesh","Himachal Pradesh")
aggregated_insurance["states"]=aggregated_insurance["states"].str.replace("Madhyapradesh","Madhya Pradesh")
aggregated_insurance["states"]=aggregated_insurance["states"].str.replace("Tamilnadu","Tamil Nadu")
aggregated_insurance["states"]=aggregated_insurance["states"].str.replace("Uttarpradesh","Uttar Pradesh")
aggregated_insurance["states"]=aggregated_insurance["states"].str.replace("Westbengal","West Bengal")
aggregated_insurance
```
